```python
import os
import json
import pandas as pd
import glob
```

```python
path = 'enter_path_of_output_folder'

file_list = os.listdir(path)

for file in file_list:
    print('filename: {}'.format(file))
```

```python
submission = pd.DataFrame(data={'patientId': [None] * len(file_list), 
                                'PredictionString': [None] * len(file_list)})
submission = submission[['patientId', 'PredictionString']]
```

```python
for i, counter1 in zip(file_list, range(len(file_list))):
    file_name = file_list[counter1]
    with open(path+'\\'+i) as file:
        data = json.load(file)
        submission.iloc[counter1]['patientId'] = file_name[0:-5]
        prediction_string = ''
        for counter2, bounding_box in enumerate(data):
            prediction_string += (str(bounding_box['confidence']) + " " + 
                                      str(bounding_box['topleft']['x']) + " " + 
                                      str(bounding_box['topleft']['y']) + " " + 
                                      str(bounding_box['bottomright']['x'] - 
                                          bounding_box['topleft']['x']) + " " +
                                      str(bounding_box['bottomright']['y'] - 
                                          bounding_box['topleft']['y']) + " ")
        submission.iloc[counter1]['PredictionString'] = prediction_string
```
